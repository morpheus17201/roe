```python
import pandas as pd
import sqlite3
```

```python
db = r"./roe4_data/violations.db"
query = """
SELECT business_id, COUNT(*) AS violation_count
FROM violations
WHERE risk_category = 'Moderate Risk'
  AND strftime('%w', date) = '1'  -- %w returns the day of the week (0 = Sunday, 1 = Monday)
GROUP BY business_id;

"""
conn = sqlite3.connect(db)
cursor = conn.cursor()
cursor.execute(query)
results = cursor.fetchall()

bus_id_list = []
for row in results:
    business_id, violation_count = row
    bus_id_list.append([business_id, violation_count])
    #print(f"Business ID: {business_id}, Violations Count: {violation_count}")

cursor.close()
conn.close()
```

```python
all_data = []
for file_num, html_file in enumerate(html_files):
    with open(html_file, "r") as f:
        text = f.read()

    
    soup = BeautifulSoup(text, "html.parser")

    tables = soup.find_all("table")
    print(f"File number: {file_num:3d}: {len(tables)=}; {html_file}", )

    for table in tables:
        # table = tables[0]
        rows = table.find_all("tr")
        dict_for_table = {}
        for row in rows:
            cells = row.find_all("td")
            if len(cells)>2:
                # print([cell.string for cell in cells])
                dict_for_table[cells[0].string] = cells[1].string
                dict_for_table[cells[2].string] = cells[3].string
        #print(dict_for_table)
        all_data.append(dict_for_table)
```
